```python
import numpy as np
import matplotlib.pyplot as plt
```

# Señales continuas con python

```python
w = 1 #frecuencia angular
fx = lambda t: np.sin(w*t) # función senoidal
t0 = 0 #tiempo inicial
tn = 2*np.pi
n = 100 #número de muestras
```

```python
dt = (tn-t0)/n #diferencia entre cada intervalo de tiempo
ti = np.arange(t0,tn,dt) #np.arange(valor_inicial, valor_final, intervalo)
senal = fx(ti)#valor de la señal por cada intervalo de tiempo
```

```python
np.set_printoptions(precision = 4)
print('tiempo: ')
print(ti)
print('señal: x(t)')
print(senal)
```

```python
plt.plot(ti,senal)
plt.title("Señal senoidal")
plt.axhline(0, color='red')
plt.axvline(0, color='red')
plt.xlabel('tiempo')
plt.ylabel('valor')
plt.show()
```

# Señales discretas


```python
w0 = 2*np.pi/19
fx_d = lambda n: np.sin(w0*n)
n0 = 0
m = 20
```

```python
n = np.arange(n0, m+1, 1)
senal_d = fx_d(n)
```

```python
print('n: ')
print(n)
print('señal x[n]: ')
print(senal_d)
```

```python
plt.stem(n, senal_d)
plt.title("Señal en tiempo discreto")
plt.xlabel('n')
plt.ylabel('señal x[n]')
plt.show()
```
